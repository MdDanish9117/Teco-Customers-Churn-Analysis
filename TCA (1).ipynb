```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Customer Churn.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"]=df["TotalCharges"].astype("float")
df.info()
```

# replacing blanks witrh 0 as tenure is 0 and no charges are recorded

```python
df["TotalCharges"]=df['TotalCharges'].replace(" ","0")
df
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

        
df['SeniorCitizen']=df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
ax=sns.countplot(x= 'Churn' , data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Churn")
plt.show()
```

```python
gb=df.groupby('Churn').agg({"Churn":'count'})
plt.pie(gb['Churn'],labels = gb.index,autopct ="%1.2f%%")
plt.title("Percentage of churned customers", fontsize=10)
plt.show()
gb
```

#from the given pie chart we can conclude that 26.54% of out customer have churned out. # not lets ecplore the reason behind it


```python
plt.figure(figsize=(4,4))
sns.countplot(x="gender",data=df , hue="Churn")
plt.title("Churn by gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="SeniorCitizen",data=df , hue="Churn")
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customer by Senior citizen")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure", data=df , bins=72, hue="Churn")
plt.show()
```
